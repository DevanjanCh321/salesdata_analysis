```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\DEBANJAN\Desktop\New folder\Diwali_Sales_data.csv', encoding='unicode_escape')

# Function to plot frequency counts for a categorical variable with enhanced visuals
def plot_frequency_counts(df, column, title, rotation=45):
    plt.figure(figsize=(12, 7))
    ax = sns.countplot(data=df, x=column, order=df[column].value_counts().index, palette='viridis')
    
    # Adding annotations
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12, color='black', weight='bold')
    
    # Customizing plot
    ax.set_title(title, fontsize=16, weight='bold')
    ax.set_xlabel(column, fontsize=14, weight='bold')
    ax.set_ylabel('Frequency', fontsize=14, weight='bold')
    plt.xticks(rotation=rotation, fontsize=12, ha='right')
    plt.yticks(fontsize=12)
    sns.despine(top=True, right=True, left=False, bottom=False)  # Removing top and right spines
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding gridlines for better readability
    
    plt.tight_layout()
    plt.show()

# Gender
plot_frequency_counts(df, 'Gender', 'Frequency Count of Gender')

# Age Group
plot_frequency_counts(df, 'Age Group', 'Frequency Count of Age Group')

# Marital Status
plot_frequency_counts(df, 'Marital_Status', 'Frequency Count of Marital Status')

# State
plot_frequency_counts(df, 'State', 'Frequency Count of State')

# Zone
plot_frequency_counts(df, 'Zone', 'Frequency Count of Zone')

# Occupation
plot_frequency_counts(df, 'Occupation', 'Frequency Count of Occupation')

# Product Category
plot_frequency_counts(df, 'Product_Category', 'Frequency Count of Product Category', rotation=90)
```

Key Performance Indicators (KPIs):


1.Total Sales Revenue
2.Average Order Value
3.Total Number of Orders
4.Number of Unique Customers
5.Top Selling Products
6.Customer Lifetime Value
7.Repeat Customer Rate

Total Sales Revenue

```python
total_sales = df['Amount'].sum()
print(f'Total Sales Revenue: Rs. {total_sales:,.2f}')
```

Average Order Value

```python
average_order_value = df['Amount'].mean()
print(f'Average Order Value: Rs. {average_order_value:,.2f}')
```

Total Number of Orders

```python
total_orders = df['Orders'].sum()
print(f'Total Number of Orders: {total_orders}')
```

Number of Unique Customers

```python
unique_customers = df['User_ID'].nunique()
print(f'Number of Unique Customers: {unique_customers}')
```

Top Selling Products

```python
top_selling_products = df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 7))
sns.barplot(y=top_selling_products.index, x=top_selling_products.values, palette='viridis')
plt.title('Top 10 Selling Products by Sales', fontsize=16, weight='bold')
plt.xlabel('Total Sales ($)', fontsize=14, weight='bold')
plt.ylabel('Product ID', fontsize=14, weight='bold')
plt.xticks(fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Customer Lifetime Value (CLV)

Customer lifetime value is a measurement of how valuable a customer is to your company, not just on a purchase-by-purchase basis but across entire customer relationships. Calculates the mean total amount spent by each unique customer.

```python
clv = df.groupby('User_ID')['Amount'].sum().mean()
print(f'Customer Lifetime Value: Rs. {clv:,.2f}')
```

Repeat Customer Rate : Calculates the percentage of customers who have made more than one purchase.

```python
repeat_customers = df[df.duplicated(subset=['User_ID'], keep=False)]
repeat_customer_rate = repeat_customers['User_ID'].nunique() / unique_customers * 100
print(f'Repeat Customer Rate: {repeat_customer_rate:.2f}%')
```

GEOGRAPHICAL ANALYSIS

```python
# State-wise Sales
state_wise_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False)
print(state_wise_sales)

plt.figure(figsize=(12, 7))
sns.barplot(x=state_wise_sales.index, y=state_wise_sales.values, palette='viridis')
plt.title('State-wise Sales', fontsize=16, weight='bold')
plt.xlabel('State', fontsize=14, weight='bold')
plt.ylabel('Total Sales ($)', fontsize=14, weight='bold')
plt.xticks(rotation=80, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Zone-wise Analysis as Pie Chart
zone_wise_sales = df.groupby('Zone')['Amount'].sum().sort_values(ascending=False)
print(zone_wise_sales)

# Plotting Pie Chart
plt.figure(figsize=(12, 7))
plt.pie(zone_wise_sales.values, labels=zone_wise_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(zone_wise_sales)))
plt.title('Zone-wise Sales', fontsize=16, weight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

CORRELATION ANALYSIS (MATRIX)

```python
# Selecting numerical columns for correlation analysis
numerical_cols = ['Age', 'Orders', 'Amount']

# Compute the correlation matrix
corr_matrix = df[numerical_cols].corr()
print(corr_matrix)

# Plot the correlation matrix
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16, weight='bold')
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.show()
```
